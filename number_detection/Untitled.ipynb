```python
from ultralytics import YOLO
import os
import numpy as np
import cv2 as cv
import sys
from matplotlib import pyplot as plt

MODEL_PATH = os.path.join('runs','detect','train6','weights','best.pt')
model = YOLO(MODEL_PATH)

img = cv.imread("datasets/test/images/8fda9fc3-6b43-11ee-bba8-fcb3bc5cd112.17.jpg")
results = model(img)
boxes = results[0].boxes.xyxy
for i in boxes:
    top = tuple([int(i[0].item()),int(i[1].item())])
    bottom = tuple([int(i[2].item()),int(i[3].item())])
    cv.rectangle(img,top,bottom, (255,0,0), 2)

plt.imshow(img)
```
